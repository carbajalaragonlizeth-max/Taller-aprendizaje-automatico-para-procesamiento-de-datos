# **Preprocesamiento de Datos**

## Limpieza de Datos

image.png

```python
print("Valores faltantes:\n", cancer_df.isnull().sum())
```

```python
# Simular valores faltantes
for i in range(300):
  cancer_df.iloc[np.random.randint(0, len(cancer_df)), np.random.randint(0, len(cancer_df.columns))] = np.nan

print(cancer_df.isnull().sum())
```

```python
# Imputación de valores por 'media'

num_cols = cancer_df.select_dtypes(include=[np.number]).columns

cancer_df_clean = cancer_df.copy()
cancer_df_clean[num_cols] = cancer_df_clean[num_cols].fillna(cancer_df_clean[num_cols].mean())

print("Después de la limpieza:\n", cancer_df_clean.isnull().sum())
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,4))
sns.histplot(cancer_df['perimeter error'], kde=True, ax=ax[0], color='red')
ax[0].set_title('radius error con valores faltantes')
sns.histplot(cancer_df_clean['perimeter error'], kde=True, ax=ax[1], color='green')
ax[1].set_title('radius error después de la imputación')
plt.show()
```

## Escalamiento

image.png

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# StandardScaler (mean=0, variance=1)
scaler_std = StandardScaler()
cancer_df_scaled = cancer_df_clean.copy()
cancer_df_scaled[num_cols] = scaler_std.fit_transform(cancer_df_clean[num_cols])

# MinMaxScaler (range: 0-1)
scaler_mm = MinMaxScaler()
cancer_df_mm = cancer_df_clean.copy()
cancer_df_mm[num_cols] = scaler_mm.fit_transform(cancer_df_clean[num_cols])

# Scatter plot para visualizar la transformación
plt.figure(figsize=(6,6))
plt.scatter(cancer_df_clean['mean area'], cancer_df_clean['worst radius'], alpha=0.5, label='Original')
plt.scatter(cancer_df_scaled['mean area'], cancer_df_scaled['worst radius'], alpha=0.5, label='Standarized')
plt.scatter(cancer_df_mm['mean area'], cancer_df_mm['worst radius'], alpha=0.5, label='MinMax')
plt.legend()
plt.xlabel('mean area')
plt.ylabel('worst radius')
plt.title('Scaling')
plt.grid(ls='--', alpha=0.5)
plt.show()
```

## Selección de características

image.png

```python
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(f_classif, k=5)
selector.fit(cancer_df_scaled, data.target)

# Graficar scores
scores = selector.scores_
plt.figure(figsize=(10,4))
sns.barplot(x=data.feature_names, y=scores, order=data.feature_names[np.argsort(scores)])
plt.xticks(rotation=90)
plt.title("Relevancia de las características (ANOVA F-score)")
plt.show()
```

```python
# Select top 5
cancer_df_top5 = cancer_df_scaled.iloc[:, selector.get_support()]

print("Forma original:", cancer_df_scaled.shape)
print("Después de selección:", cancer_df_top5.shape)
```

## Extracción de Características

image.png

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

cancer_df_PCA = pca.fit_transform(pd.get_dummies(cancer_df_top5))

plt.figure(figsize=(6,6))
plt.scatter(cancer_df_PCA[:,0], cancer_df_PCA[:,1], c=data.target, cmap='coolwarm', alpha=0.6)
plt.title('PCA con 2 componentes')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar(label='Cancer (0 = No, 1 = Yes)')
plt.grid(ls='--', alpha=0.5)
plt.show()
```

# **Ejemplo completo**

## **1.** Generando el Dataset

```python
# Fijamos semilla para que a todos los alumnos les dé exactamente el mismo error
np.random.seed(12345)

n = 1000
ciudades = ['Mérida', 'CDMX', 'Monterrey']

df = pd.DataFrame({
    'edad': np.random.normal(35, 10, n).astype(int),
    'ingresos': np.random.exponential(20000, n),
    'monto_prestamo': np.random.normal(150000, 50000, n),
    'ciudad': np.random.choice(ciudades, n),
    # Variable ruidosa 1: Completamente aleatoria (No sirve para predecir)
    'ruido_puro': np.random.normal(0, 1, n),
    # Variable ruidosa 2: Una copia multiplicada de la edad (Multicolinealidad)
    'edad_falsa_redundante': np.zeros(n)
})

df
```

```python
df['edad_falsa_redundante'] = df['edad'] * 1.8 + np.random.normal(0, 2, n)

df
```

```python
# Inyectamos el Target (Ligeramente dependiente de ingresos y monto)
prob_impago = 1 / (1 + np.exp(-(-2 + (df['monto_prestamo']/100000) - (df['ingresos']/30000))))
df['target'] = np.random.binomial(1, p=prob_impago)

df
```

```python
# (Corrompemos el dataset)
# 1. Metemos NaNs en el 10% de los ingresos
df.loc[df.sample(frac=0.1, random_state=1).index, 'ingresos'] = np.nan

df
```

```python
from sklearn.model_selection import train_test_split

# Dividimos en Características y Targets
X = df.drop(columns=['target'])
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12345)

print(f"Dimensión de Entrenamiento: {X_train.shape} | Dimensión de Test: {X_test.shape}")
print(f"Dimensión de Entrenamiento: {y_train.shape} | Dimensión de Test: {y_test.shape}")
```

## **2.** Limpieza y Escalado

### Tratamiento de NaNs (Imputación)
Usaremos la Mediana en lugar de la Media para los ingresos, porque la Media es arrastrada por los millonarios.

### Codificación Categórica (One-Hot Encoding)
Convertimos "Mérida" en [1, 0]. Usamos drop='first' para evitar la Trampa de las Variables Ficticias (si no eres de CDMX ni de Monterrey, matemáticamente se infiere que eres de Mérida; la tercera columna sobra).

image.png

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

# Definimos tipos de variables
vars_numericas = ['edad', 'ingresos', 'monto_prestamo', 'ruido_puro', 'edad_falsa_redundante']
vars_categoricas = ['ciudad']

# Definimos las codificaciones para cada tipo de variable
config_numerica = SimpleImputer(strategy='median') #Tranformar las caracteristicas mediante la imputación
config_categorica = OneHotEncoder(drop='first', sparse_output=False) #Podemos eliminar una

# Empaquetamos todo en un procesador maestro
preprocesador = ColumnTransformer(
    transformers=[
        ('num', config_numerica, vars_numericas),
        ('cat', config_categorica, vars_categoricas)
    ],
    remainder='passthrough'
)

# APRENDEMOS la matemática solo del Train, pero TRANSFORMAMOS ambos
X_train_limpio = preprocesador.fit_transform(X_train)
X_test_limpio = preprocesador.transform(X_test)

# Recuperamos los nombres de las columnas generadas
nombres_cat = preprocesador.named_transformers_['cat'].get_feature_names_out(vars_categoricas)
nombres_finales = vars_numericas + list(nombres_cat)

df_train_limpio = pd.DataFrame(X_train_limpio, columns=nombres_finales)
df_test_limpio = pd.DataFrame(X_test_limpio, columns=nombres_finales)

print("Previo a la Limpieza")
print("NaNs en Train:", X_train.isnull().sum().sum())
print("NaNs en Test:", X_test.isnull().sum().sum())
print("Después de la Limpieza")
print("NaNs en Train:", df_train_limpio.isnull().sum().sum())
print("NaNs en Test:", df_test_limpio.isnull().sum().sum())
```

### Contención de Outliers (Clipping por IQR) y Escalado
La fila 0 del df_train_limpio: tenemos a un sujeto de 999 años. Si se ingresa a una regresión, genera ruido.

```python
# Metemos un Outlier ridículo en la fila 0
df_train_limpio.loc[0, 'edad'] = 999
df_train_limpio.head(3)
```

Aplicaremos un Clipping: a todo el que pase del límite superior del Rango Intercuartílico ($Q_3 + 1.5 \times IQR$), lo "toparemos" obligatoriamente en ese límite máximo.


```python
# Función para topar Outliers basada estrictamente en el Train
def topar_outliers(df_train, df_test, columna):
    Q1 = df_train[columna].quantile(0.25)
    Q3 = df_train[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_sup = Q3 + 1.5 * IQR

    print('Q1:', Q1)
    print('Q3:', Q3)
    print('IQR:', IQR)
    print('Límite Superior:', limite_sup)

    # Aplicamos el tope calculado del train a ambos datasets
    df_train[columna] = np.clip(df_train[columna], a_min=None, a_max=limite_sup)
    df_test[columna] = np.clip(df_test[columna], a_min=None, a_max=limite_sup)
    return df_train, df_test

# Topamos la edad
df_train_limpio, df_test_limpio = topar_outliers(df_train_limpio, df_test_limpio, 'edad')
```

Posterior a eso, aplicamos un StandardScaler ($\mu$=0, $\sigma$=1).

```python
# Aplicamos StandardScaler a todo
escalador = StandardScaler()
X_train_escalado = escalador.fit_transform(df_train_limpio)
X_test_escalado = escalador.transform(df_test_limpio)

# Reconstruimos los DataFrames listos
df_train_listo = pd.DataFrame(X_train_escalado, columns=nombres_finales)
print("Edad máxima ahora en el dataset:", df_train_listo['edad'].max())
```

## **3.** Reducción de la Dimensionalidad

### Destilación y Selección de Características
¿Por qué menos es más? La Maldición de la Dimensionalidad (Basado en Murphy, Capítulo 8).
En 1 dimensión (una línea de 10 metros), 10 puntos están muy apretados. En 2 dimensiones (un suelo de 10mx10m), 10 puntos ya están algo dispersos. En 100 dimensiones, el volumen del hiperespacio crece tan rápido que los datos quedan a años luz unos de otros.
Cuando el espacio está vacío, el algoritmo de Machine Learning no encuentra patrones; se asusta y decide memorizarse de memoria la posición exacta de cada punto (Overfitting). Para curar esto, tenemos dos caminos: Comprimir (Extracción) o Podar (Selección).

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Evaluamos todas las columnas contra la variable Target
selector_filtro = SelectKBest(score_func=f_classif, k='all')
selector_filtro.fit(df_train_listo, y_train)

# Armamos una tabla de posiciones
tabla_filtro = pd.DataFrame({
    'Variable': df_train_listo.columns,
    'F-Score (Poder predictivo)': selector_filtro.scores_
}).sort_values(by='F-Score (Poder predictivo)', ascending=False)

print("Ranking por el método de filtrado ANOVA")
print(tabla_filtro.to_string(index=False))
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(x='Variable', y='F-Score (Poder predictivo)', data=tabla_filtro, palette='viridis')
plt.xticks(rotation=90)
plt.title('Ranking de Características por ANOVA')
plt.show()
```

### Extracción de Características
Análisis de Componentes Principales (PCA)
PCA busca el ángulo en el hiperespacio donde los datos están más "estirados" (mayor varianza). En lugar de pedirle k=3 componentes a ciegas, le pediremos a Python: «Dame el número mínimo de componentes necesarias para retener el 80% de la información original».

```python
from sklearn.decomposition import PCA

# Le pedimos el 80% de la varianza explicada
pca = PCA(n_components=0.80)

X_train_pca = pca.fit_transform(X_train_escalado)

print(f"Teníamos {X_train_escalado.shape[1]} columnas originales.")
print(f"PCA logró comprimir el 80% de la información en solo {X_train_pca.shape[1]} componentes matemáticas.")
print("Varianza explicada por cada componente:", pca.explained_variance_ratio_.round(3))
```

## **4.** Métodos Embebidos (Embedded Methods)

### Aquí se emplea un algoritmo que se entrena:

* **Importancia por Random Forest**: El bosque cuenta cuántas veces usó una variable para partir un nodo del árbol.

* **Regularización LASSO (L1)**: Es un algoritmo de regresión al que le cobramos un "impuesto" por cada variable que use. Para ahorrar impuestos, Lasso multiplica por cero los coeficientes de las variables que no aportan mucha información.

image.png

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# 1. Aplicamos LASSO (penalización L1)
lasso = LogisticRegression(penalty='l1', solver='liblinear', C=0.1, random_state=12345)
lasso.fit(df_train_listo, y_train)

# 2. Aplicamos Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=12345)
rf.fit(df_train_listo, y_train)

tabla_veredicto = pd.DataFrame({
    'Atributo': df_train_listo.columns,
    'Coeficiente LASSO (Cero = Eliminada)': lasso.coef_[0].round(4),
    'Importancia Random Forest (%)': (rf.feature_importances_ * 100).round(2)
}).sort_values(by='Importancia Random Forest (%)', ascending=False)

print("Auditoría de Métodos Embebidos")
print(tabla_veredicto.to_string(index=False))
```
