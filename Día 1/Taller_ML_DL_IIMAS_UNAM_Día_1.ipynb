## ¿Qué es Google Colab y cómo funciona por dentro?
**Google Colaboratory (o Colab)** no es una aplicación web de edición de texto común; es una interfaz basada en la nube para **Jupyter Notebooks**.

---
Cuando se abre este notebook, ocurre una secuencia de eventos:
*   **Asignación de una Máquina Virtual (VM)**: Google asigna de forma gratuita una VM corriendo en una distribución de **Linux (Ubuntu)** dentro de sus centros de datos.
*   **El Servidor Jupyter**: Dentro de esa VM, corre un servidor de **Python** activo que está escuchando las órdenes que se envían desde el navegador web.
*  **El Kernel**: Cada celda de código que se ejecuta es enviada a un motor de ejecución (**Kernel de Python**). Este procesa las instrucciones, altera el estado de la memoria de la VM y devuelve la respuesta.

# **Bondades de Colab en su utilización:**

*  ## Zero Setup (Configuración Cero)
    Instalar librerías de Machine Learning en entornos locales (especialmente con soporte para GPU como CUDA) suele generar dificultad por dependencias del sistema operativo. Colab viene con el 95% de las librerías científicas preinstaladas (`NumPy`, `Pandas`, `Scikit-Learn`, `PyTorch`, `TensorFlow`).
*  ## El paradigma REPL y la Persistencia de Memoria en Caliente
    A diferencia del desarrollo de software tradicional donde se escribe un script entero, se compila y se corre de principio a fin, la Ciencia de Datos requiere **exploración iterativa**.
    Los notebooks implementan un bucle **REPL (*Read-Eval-Print Loop*)**. Esto significa que la memoria persiste entre celdas. Se puede cargar un conjunto de datos pesado en la Celda 1 (lo cual toma tiempo) y pasar las próximas 3 horas limpiando y analizando esos datos en la Celda 2 sin tener que volver a recargar el archivo desde el disco cada vez.
*  ## Almacenamiento e Integración con el Ecosistema
    Al estar integrado con **Google Drive**, no depende del almacenamiento físico de la máquina física. Se puede montar el Drive directamente como si fuera una carpeta local del servidor Linux usando código, lo que facilita la lectura de datasets gigantes y el guardado de los pesos de los modelos entrenados.
*  ## IMPORTANTE⚠️: La VM es efímera
    Si se deja el notebook inactivo por mucho tiempo o se supera el límite de horas continuas de la sesión gratuita, Google desconectará el contenedor y borrará todo lo que esté guardado localmente en el disco de la VM.

## Ejemplo de REPL

```python
# NumPy es una biblioteca para el lenguaje de programación Python que da soporte
# para crear vectores y matrices grandes multidimensionales, junto con una gran
# colección de funciones matemáticas de alto nivel para operar con ellas.
import numpy as np
```

```python
# Celda A: Se define una matriz de datos pesada y simulamos su carga en memoria

print("Cargando matriz de datos a la memoria de la VM...")
datos_clientes = np.random.normal(loc=50, scale=10, size=(100000, 5))
print("Matriz cargada: está en la RAM.")
```

```python
# Celda B: Se opera sobre la variable 'datos_clientes' creada arriba sin recargarla.

media_por_columna = datos_clientes.mean(axis=0)
print("La media calculada directamente desde la RAM persistente es:")
print(media_por_columna)
```
